```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Cargar los datos
file_path = '/home/elicoubuntu/Desktop/Industrial_Insigths/Prognosis/consolidated_data.xlsx'
data = pd.read_excel(file_path)

# Seleccionar solo las columnas numéricas
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Normalizar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)

# Aplicar el clustering con K-means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_data)

# Asignar los clústeres al dataset original
data['Cluster'] = kmeans.labels_

# Visualización básica (opcional)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('Clustering de Datos Eléctricos')
plt.show()

# Guardar los datos con los clústeres asignados
data.to_excel('/home/elicoubuntu/Desktop/Industrial_Insigths/Prognosis/dataset_con_clustering.xlsx', index=False)
```
